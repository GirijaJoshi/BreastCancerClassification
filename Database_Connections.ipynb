# Find kaggle image files and store them in the Postgres Database BreastCancer Classification

```python
#Find our data set by name search
datasets=api.dataset_list(search='breast-histopathology-images')
print(datasets)
```

```python
# store our data set
Our_set = datasets[0]
# Store the image file names in a dictionary
read_image_file_dict = {'file_path_names':api.dataset_list_files(f'{Our_set}').files}
#Create data frame from the dictionary
image_file_df = pd.DataFrame.from_dict(read_image_file_dict)
image_file_df.head()
```

```python
image_file_df.to_sql(name='Kaggle_Images', con=engine)
```
